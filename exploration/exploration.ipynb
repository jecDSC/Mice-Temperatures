```python
import pandas as pd
import numpy as np
```

```python
female = pd.read_csv('..\docs\dataset\Mouse_Data_Student_Copy.xlsx - Fem Temp.csv')
female
```

```python
male = pd.read_csv('..\docs\dataset\Mouse_Data_Student_Copy.xlsx - Male Temp.csv')
male
```

```python
male_df = pd.DataFrame({'temp': male.mean(axis = 1), 'sex': 'male'}).reset_index(names = 'min')
```

```python
female_df = pd.DataFrame({'temp': female.mean(axis = 1), 'sex': 'female'}).reset_index(names = 'min')
```

```python
female_df
```

```python
male_df
```

```python
df = pd.concat([male_df, female_df], ignore_index=True)
df = df[df['min'] <= 8640]
df.to_csv('../docs/dataset/average_temp.csv')
```

```python
diff = male_df[['min']]
diff['diff'] = female_df['temp'] - male_df['temp']
diff = diff[diff['min'] <= 8640]
diff.to_csv('../docs/dataset/male_female_difference.csv')
```
