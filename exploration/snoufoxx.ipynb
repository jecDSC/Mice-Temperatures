```python
avgsm = []
avgsf = []
for i in range(0, 20160, 60):
    summ = 0
    sumf = 0
    for j in range(13):
        summ += micemt[int(f'{i}'):int(f'{i+60}')][f'm{j + 1}'].mean()
        sumf += miceft[int(f'{i}'):int(f'{i+60}')][f'f{j + 1}'].mean()
    avgsm.append(summ / 13)
    avgsf.append(sumf / 13)
```

```python
avgsall = pd.DataFrame()
avgsall['Male'] = avgsm
avgsall['Female'] = avgsf
avgsall.shape
```

```python
fig1 = px.line(avgsall, title='Hourly Core Temp. Averages for Male/Female Mice')
fig1.show()
```

```python
avgsmmin = []
avgsfmin = []
for i in range(10000):
    avgsmmin.append(micemt.iloc[i].mean())
    avgsfmin.append(miceft.iloc[i].mean())

avgsminf = pd.DataFrame()
avgsminf['Temp.'] = avgsmmin
avgsminf['Light_On'] = on
avgsminf['Sex'] = ["female"] * avgsminf.shape[0]
avgsminf = avgsminf[:10000]

avgsminm = pd.DataFrame()
avgsminm['Temp.'] = avgsfmin
avgsminm['Light_On'] = on
avgsminm['Sex'] = ["male"] * avgsminf.shape[0]
avgsminm = avgsminm[:10000]
```
